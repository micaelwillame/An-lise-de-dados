```python
# removendo as linhas duplicadas da tabela
df.drop_duplicates(inplace=True)
```

```python
# subistituindo os '/' das colunas por '-' e convertendo para um formato de data e hora, assumindo que o formato é dia-mês-ano.
df['order_date'] = df['order_date'].str.replace('/','-')
df['order_date'] = pd.to_datetime(df['order_date'],format='%d-%m-%Y')
df['ship_date'] = df['ship_date'].str.replace('/','-')
df['ship_date'] = pd.to_datetime(df['ship_date'],format='%d-%m-%Y')
```

```python
# extraindo e criando colunas com o ano, mês, dia da coluna order.
# extraindo o dia da semana (onde segunda-feira é 0 e domingo é 6) de 'order_date' e cria uma nova coluna 'order_weekday'
# calculando a diferença em dias entre 'ship_date' e 'order_date' e cria uma nova coluna 'ship_days'.
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_day'] = df['order_date'].dt.day
df['order_weekday'] = df['order_date'].dt.weekday+1
df['order_year_month'] = df['order_date'].dt.strftime('%Y-%m')
df['ship_days'] = (df['ship_date']-df['order_date']).dt.days
```

```python
# removendo vírgulas da coluna 'sales' e converte os valores para números float.
# calculando o custo subtraindo a coluna 'profit' da coluna 'sales'.
df['sales'] = df['sales'].str.replace(',','').astype(float)
df['cost'] = df['sales'] - df['profit']
```

```python
# agrupando o DataFrame df pela coluna 'market' e, em seguida, calcula algumas estatísticas agregadas para cada grupo.
# ele soma as vendas ('sales'), conta a quantidade única de pedidos ('order_id'), e soma os lucros ('profit') para cada mercado.
df_market = df.groupby('market').agg({'sales':'sum','order_id':'nunique','profit':'sum'})
```

```python
# calculando a média de vendas por pedido para cada mercado e
# adicionando essa informação como uma nova coluna chamada 'average_sales' no DataFrame df_market.
df_market['average_sales'] = df_market['sales']/df_market['order_id']
```

```python
# calculando a taxa de lucro (lucro dividido por vendas) para cada mercado e
# adicionando essa informação como uma nova coluna chamada 'profit_rate' no DataFrame df_market.
df_market['profit_rate'] = df_market['profit']/df_market['sales']
```

```python
# renomeia a coluna 'order_id' para 'order_num' no DataFrame df_market.
# isso é feito usando o método rename e especificando um dicionário para mapear os nomes das colunas.
df_market.rename(columns={'order_id':'order_num'},inplace=True)
```

```python
import plotly.subplots as sp
import plotly.graph_objects as go

# Resetando índices
fig_df = df_market.reset_index()

# Criando subplots
fig = sp.make_subplots(rows=2, cols=2, subplot_titles=['Sales', 'Average Sales', 'Order Number', 'Profit Rate'])

# Adicionando gráficos de barras para cada métrica
for i, metric in enumerate(['sales', 'average_sales', 'order_num', 'profit_rate'], 1):
    row = (i - 1) // 2 + 1
    col = (i - 1) % 2 + 1
    fig.add_trace(go.Bar(
        x=fig_df['market'],
        y=fig_df[metric],
        name=metric,
    ), row=row, col=col)

# Atualizando layout
fig.update_layout(
    title_text='Metrics by Market',
    showlegend=False,
)

# Exibindo os subplots
fig.show()
```
